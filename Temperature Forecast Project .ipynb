```python
# Convert 'Date' column to datetime format
temp_data['Date'] = pd.to_datetime(temp_data['Date'])
```

```python
from sklearn.impute import SimpleImputer

# Define columns with missing values
columns_with_missing = ['Present_Tmax', 'Present_Tmin', 'LDAPS_RHmin', 'LDAPS_RHmax',
                        'LDAPS_Tmax_lapse', 'LDAPS_Tmin_lapse', 'LDAPS_WS', 'LDAPS_LH',
                        'LDAPS_CC1', 'LDAPS_CC2', 'LDAPS_CC3', 'LDAPS_CC4', 'LDAPS_PPT1',
                        'LDAPS_PPT2', 'LDAPS_PPT3', 'LDAPS_PPT4', 'Next_Tmax', 'Next_Tmin']

# Impute missing values with mean for each column
for column in columns_with_missing:
    imputer = SimpleImputer(strategy='mean')
    temp_data[column] = imputer.fit_transform(temp_data[[column]])

float_columns = ['station', 'Present_Tmax', 'Present_Tmin', 'LDAPS_RHmin', 'LDAPS_RHmax',
                 'LDAPS_Tmax_lapse', 'LDAPS_Tmin_lapse', 'LDAPS_WS', 'LDAPS_LH',
                 'LDAPS_CC1', 'LDAPS_CC2', 'LDAPS_CC3', 'LDAPS_CC4', 'LDAPS_PPT1',
                 'LDAPS_PPT2', 'LDAPS_PPT3', 'LDAPS_PPT4', 'lat', 'lon', 'DEM',
                 'Slope', 'Solar radiation', 'Next_Tmax', 'Next_Tmin']

# Convert columns to appropriate data types
for column in float_columns:
    temp_data[column] = temp_data[column].astype(float)

print(temp_data.info())
print(temp_data.isnull().sum())
```

```python
from sklearn.model_selection import train_test_split

# Define features (X) and target variables (y)
X = temp_data.drop(['Next_Tmax', 'Next_Tmin', 'Date'], axis=1)  # Drop 'Date' as it's not a numerical feature
y_max = temp_data['Next_Tmax']
y_min = temp_data['Next_Tmin']

# Split the data into training and test sets (80% training, 20% test)
X_train, X_test, y_max_train, y_max_test = train_test_split(X, y_max, test_size=0.2, random_state=42)
_, _, y_min_train, y_min_test = train_test_split(X, y_min, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# Initialize the StandardScaler
scaler = StandardScaler()

# Fit the scaler on the training data and transform both training and test data
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
```

```python
models = {
    'RandomForest': RandomForestRegressor(),
    'LinearRegression': LinearRegression(),
    'SVR': SVR(),
    'KNeighbors': KNeighborsRegressor(),
}

for name, model in models.items():
    print(f"Training and evaluating {name}...")
    
    # Perform cross-validation
    cv_scores = cross_val_score(model, X_scaled, y_max, cv=5, scoring='neg_mean_squared_error')
    
    # Calculate scores in percentage
    train_score = model.fit(X_train_scaled, y_max_train).score(X_train_scaled, y_max_train) * 100
    test_score = model.score(X_test_scaled, y_max_test) * 100
    
    print(f"Cross-Validation MSE for {name}: {np.mean(cv_scores)}")
    print(f"Training Score for {name}: {train_score:.2f}%")
    print(f"Test Score for {name}: {test_score:.2f}%\n")
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'n_neighbors': [3, 5, 7, 9, 11],  
    'weights': ['uniform', 'distance'],
    'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute']
}
```

```python
# Initialize KNeighborsRegressor
knn = KNeighborsRegressor()

# Initialize GridSearchCV
grid_search = GridSearchCV(knn, param_grid, cv=5, scoring='neg_mean_squared_error', verbose=1, n_jobs=-1)

# Fit GridSearchCV
grid_search.fit(X_scaled, y_max) 
```

```python
# Best hyperparameters
best_params = grid_search.best_params_
print(f"Best Hyperparameters: {best_params}")

# Best cross-validation score 
best_score = np.abs(grid_search.best_score_)  
print(f"Best Mean Squared Error: {best_score}")
```

```python
best_knn = KNeighborsRegressor(**best_params)  
best_knn.fit(X_train_scaled, y_max_train)

# Evaluate on test set
test_score = best_knn.score(X_test_scaled, y_max_test)
print(f"Test Score with Best KNeighborsRegressor: {test_score * 100:.2f}%")
```

```python
# Make predictions using the best-tuned KNeighborsRegressor
y_max_pred = best_knn.predict(X_test_scaled)
```
