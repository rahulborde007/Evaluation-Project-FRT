```python
from sklearn.preprocessing import LabelEncoder

# Initialize the label encoder
label_encoder = LabelEncoder()

df['type_encoded'] = label_encoder.fit_transform(df['type'])

print(df[['type', 'type_encoded']].head())
```

```python
from sklearn.model_selection import train_test_split

# For classification
X_cls = df[['AveragePrice', 'Total Volume', '4046', '4225', '4770', 'type_encoded']]
y_cls = df['region']
X_train_cls, X_test_cls, y_train_cls, y_test_cls = train_test_split(X_cls, y_cls, test_size=0.2, random_state=42)

# For regression
X_reg = df[['Total Volume', '4046', '4225', '4770', 'type_encoded', 'region']]  
y_reg = df['AveragePrice']
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X_reg, y_reg, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# Initialize and train the classifier
clf = RandomForestClassifier()
clf.fit(X_train_cls, y_train_cls)

# Predictions and Evaluation
y_pred_cls = clf.predict(X_test_cls)
print("Classification Accuracy:", accuracy_score(y_test_cls, y_pred_cls))
print(classification_report(y_test_cls, y_pred_cls))
```

```python
# Use get_dummies function for One-Hot Encoding
df_encoded = pd.get_dummies(df, columns=['type', 'region'], drop_first=True)
```

```python
# Features for regression with encoded 'type' and 'Region'
X_reg_encoded = df_encoded.drop(columns=['AveragePrice', 'Date']) 
y_reg_encoded = df_encoded['AveragePrice']

X_train_reg_encoded, X_test_reg_encoded, y_train_reg_encoded, y_test_reg_encoded = train_test_split(X_reg_encoded, y_reg_encoded, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Initialize and train the RandomForestRegressor
regressor = RandomForestRegressor()
regressor.fit(X_train_reg_encoded, y_train_reg_encoded)

# Predict on the test set
y_pred_reg_encoded = regressor.predict(X_test_reg_encoded)

# Evaluate the model
mse = mean_squared_error(y_test_reg_encoded, y_pred_reg_encoded)
r2 = r2_score(y_test_reg_encoded, y_pred_reg_encoded)

print(f"Mean Squared Error (MSE): {mse}")
print(f"R-squared (R2): {r2}")
```

```python
from sklearn.ensemble import RandomForestRegressor

# Initialize the RandomForestRegressor
regressor = RandomForestRegressor(random_state=42)

# Train the model
regressor.fit(X_train_reg_encoded, y_train_reg_encoded)
```

```python
from sklearn.metrics import r2_score

# Predict on the training set
y_train_pred = regressor.predict(X_train_reg_encoded)

y_test_pred = regressor.predict(X_test_reg_encoded)

train_score = r2_score(y_train_reg_encoded, y_train_pred)

test_score = r2_score(y_test_reg_encoded, y_test_pred)

train_score_percentage = train_score * 100
test_score_percentage = test_score * 100

print(f"Train Score (R-squared): {train_score_percentage:.2f}%")
print(f"Test Score (R-squared): {test_score_percentage:.2f}%")
```

```python
from sklearn.metrics import mean_squared_error

# Calculate MSE for training set
mse_train = mean_squared_error(y_train_reg_encoded, y_train_pred)

# Calculate MSE for test set
mse_test = mean_squared_error(y_test_reg_encoded, y_test_pred)

print(f"Mean Squared Error (MSE) for Training Set: {mse_train:.4f}")
print(f"Mean Squared Error (MSE) for Test Set: {mse_test:.4f}")
```

```python
from sklearn.model_selection import cross_val_score

# Perform 5-fold cross-validation
cv_scores = cross_val_score(regressor, X_reg_encoded, y_reg_encoded, cv=5, scoring='r2')

print(f"Cross-validated R-squared scores: {cv_scores}")
print(f"Average 5-Fold R-squared: {cv_scores.mean():.2f}")
```
